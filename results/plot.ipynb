```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
maxent_norevisit_test=np.load('./active_no_revisit/test_accuracy_list.npy')
maxent_norevisit_train=np.load('./active_no_revisit/train_accuracy_list.npy')
maxent_revisit_test=np.load('./active_revisit/test_accuracy_list.npy')
maxent_revisit_train=np.load('./active_revisit/train_accuracy_list.npy')
random_test=np.load('./random/test_accuracy_list.npy')
random_train=np.load('./random/train_accuracy_list.npy')

sns.set(style="whitegrid")
plt.title('test accuracy')
plt.plot(random_test,label='random')
plt.plot(maxent_norevisit_test,label='maxent_norevisit')
plt.plot(maxent_revisit_test,label='maxent_revisit')
plt.legend()
plt.show()

plt.title('train accuracy')
plt.plot(random_train,label='random')
plt.plot(maxent_norevisit_train,label='maxent_norevisit')
plt.plot(maxent_revisit_train,label='maxent_revisit')
plt.legend()
plt.show()
```
